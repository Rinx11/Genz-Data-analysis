```python
#Analyzing the factors of influence #counting values of selected column using value_counts and creating piechart
Genz['Influenced by'].value_counts().plot(kind='pie',figsize=(6,6),title="Influencial factors",colors=['deeppink','darkmagenta','darkorchid','palevioletred','rosybrown']) 
```

```python
fig,axs=plt.subplots(2,1,figsize=(10,8)) #creating subplots imp to mention axs,size in subplots() and define axis by ax=axs[0] in plots
Genz.loc[Genz['Gender'] == 'Female', 'Higher Education Aspirations Outside India '].value_counts().plot(kind='barh',title="PG outside India when gender is female",ax= axs[0],color=['deeppink','darkmagenta','darkorchid','palevioletred','rosybrown'])
Genz.loc[Genz['Gender'] == 'Male', 'Higher Education Aspirations Outside India '].value_counts().plot(kind='barh',title='PG outside India when gender is male',ax=axs[1],color=['deeppink','darkmagenta','darkorchid','palevioletred','rosybrown'])
plt.tight_layout() #for space between two plots
plt.show()
```
